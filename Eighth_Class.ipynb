```python
from datetime import datetime

today = datetime.now()
age_delta = today - birthday

age_in_years = age_delta.days / 365.25  # Account for leap years

print(f"Today's date: {today.strftime('%Y-%m-%d')}")
print(f"Birthday: {birthday.strftime('%Y-%m-%d')}")
print(f"Age: {int(age_in_years)} years")
```

```python
import pandas as pd

students = pd.DataFrame({
    "StudentID":[101,102,103,104,105,106,107,108],
    "Name":["Alice","Bob","Carol","David","Eva","Frank","Grace","Helen"],
    "Age":[22,23,21,24,20,22,23,21],
    "CGPA":[3.80,3.65,3.92,3.55,3.88,3.40,3.76,3.95],
    "Study_Hours":[6,4,8,5,7,3,6,9],
    "AI_Literacy":[88,76,95,82,91,70,86,98],
    "Digital_Literacy":[85,74,94,80,92,72,84,97],
    "Research_Productivity":[12,7,18,10,16,5,13,20]})

print(students)
```

```python
students.describe()
```

```python
students["Study_Hours"].corr(students["CGPA"])
```

```python
students.corr(numeric_only=True)
```

```python
students[["CGPA","Study_Hours","AI_Literacy"]].corr()
```

```python
#Z score

students["CGPA_Z"]=(students["CGPA"]-students["CGPA"].mean())/students["CGPA"].std()
print(students)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

students = pd.DataFrame({
    "Student":["Alice","Bob","Carol","David","Eva","Frank","Grace","Helen"],
    "CGPA":[3.80,3.65,3.92,3.55,3.88,3.40,3.76,3.95],
    "Study_Hours":[6,4,8,5,7,3,6,9],
    "AI_Literacy":[88,76,95,82,91,70,86,98],
    "Research_Papers":[12,7,18,10,16,5,13,20]
})

print(students)
```
