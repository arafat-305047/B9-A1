```python
marks = pd.Series(
    [82, 76, 91, 88, 79],
    index=["Acon","Bcon","Ccon","Dcon","Econ"]
)

print("student marks")
print(marks)

print("\nAverage marks")
print(marks.mean())

print("\nHigest marks")
print(marks.max())

print("\nlowest marks")
print(marks.min())

print("\nStandard Deviation of marks:")
print(marks.std())

print("\nMarks sorted in ascending order:")
print(marks.sort_values(ascending=True))

print("\nMarks greater than 80:")
print(marks[marks > 80])   #idxmix(), idxmin()
```
